```python
cptac.download(dataset="Ccrcc")
ccrcc = cptac.Ccrcc()
ccrcc.list_data()
```

```python
# Below, we get the clinical dataset from the ccrcc dataframe. I also preemptively removed the na's from the clinical dataset based only on tumor_stage_pathological.
```

```python
clinical = ccrcc.get_clinical()
clinical.columns = clinical.columns.get_level_values(0)

masked_clinical = clinical[~pd.isna(clinical.tumor_stage_pathological)]
masked_clinical
```

```python
# getting the protein data and rna data from the ccrcc dataset. I also removed columns that contain NA here so that data analyzed is only full columns of expression values. 
```

```python
protein_data = ccrcc.get_proteomics()
protein_data.columns = protein_data.columns.get_level_values(0)
protein_data = protein_data.dropna(axis=1)
```

```python
rna_data = ccrcc.get_transcriptomics()
rna_data.columns = rna_data.columns.get_level_values(0)
#rna_data = np.log2(rna_data)
rna_data = rna_data.replace(0, np.nan)
rna_data = np.log2(rna_data)
rna_data = rna_data.dropna(axis=1)
```

```python
# Question 1: Protein

# 1. make dataframes for protein data for stage i and stage iii patients
# 2. take the column means for each protein
# 3. find the 5 most expressed ones using pd 

# PROTEINS
stagei = masked_clinical[masked_clinical.tumor_stage_pathological == "Stage I"]
stageiii = masked_clinical[masked_clinical.tumor_stage_pathological == "Stage III"]

stagei = masked_clinical[masked_clinical.tumor_stage_pathological == "Stage I"].index
stageiii = masked_clinical[masked_clinical.tumor_stage_pathological == "Stage III"].index

stagei_shared = np.intersect1d(stagei, protein_data.index)
stageiii_shared = np.intersect1d(stageiii, protein_data.index)

stagei_protein = protein_data.loc[stagei_shared]
stageiii_protein = protein_data.loc[stageiii_shared]

mean_stagei = np.mean(stagei_protein, axis=0)
mean_stageiii = np.mean(stageiii_protein, axis=0)

df_prot = pd.DataFrame({
    'protein name': protein_data.columns,
    'difference': abs(mean_stagei - mean_stageiii), # difference in means
})

df_prot.sort_values(by='difference', ascending=False)[:5] # top 5 genes wtih absolute difference in expression for proteins
# prints LDB3, BTBD7, GDF6, COX4I2, and SNCB

top5prot = ['FTL', 'HBZ', 'HBA2', 'CMA1', 'HBB']
protein_data[top5prot]
```

```python
# Question 1: RNA
stagei_shared_rna = np.intersect1d(stagei, rna_data.index)
stageiii_shared_rna = np.intersect1d(stageiii, rna_data.index)

stagei_rna = rna_data.loc[stagei_shared_rna]
stageiii_rna = rna_data.loc[stageiii_shared_rna]

mean_stagei_rna = np.mean(stagei_rna, axis=0)
mean_stageiii_rna = np.mean(stageiii_rna, axis=0)

df_rna = pd.DataFrame({
    'gene name': rna_data.columns,
    'difference': abs(mean_stagei_rna - mean_stageiii_rna)
})

df_rna
df_rna.sort_values(by='difference', ascending=False)[:5]
# prints MT-CO1, GAPDH, TGFBI, SPP1, and VIM

top5rna = ['AJAP1', 'DPEP1', 'GALNT5', 'IL20RB', 'RYR2']
```

```python
#2: Custom dataframe
shared_protein_rna = np.intersect1d(protein_data.index, rna_data.index)
rna_protein_clinical = np.intersect1d(shared_protein_rna, masked_clinical.index)
#preemptively removes all NA from clinical

rna = rna_data[top5rna].loc[rna_protein_clinical]
protein = protein_data.loc[rna_protein_clinical]

#protein.drop(protein.columns[[4]], axis=1)
X_data = pd.concat([protein, rna], axis=1)
X_data # dataset contains patient names, and top 5 genes from RNA and protein based on difference in means
```

```python
#3: y_data is just various stages of tumor
y_data = masked_clinical["tumor_stage_pathological"]
```

```python
#4: Scaling and Encoding
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder

scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_data)

#Encoding is not necessary here because the expression values are only numeric
```

```python
#5: Training the dataset at first just to test it out
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y_data, train_size=0.7)
```

```python
#6: Comparing classification model accuracy
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB

classifier_names = [
    'KNeighborsClassifier', 
    'DecisionTreeClassifier', 
    'MLPClassifier', 
    'GaussianNB'
]

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    MLPClassifier(),
    GaussianNB(),
]

classifier_temp = {
    0: [],
    1: [],
    2: [],
    3: []
}
for n in range(10):
    X_train, X_test, y_train, y_test = train_test_split(scaled_data, y_data, train_size=0.7)
    for i in range(len(classifiers)):
        model = classifiers[i]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
        difference = sum(y_pred==y_test) / len(y_test)
        classifier_temp[i].append(difference)

```

```python
#7: The most accurate in 10 trials is the KNeighborsClassifier, followed by the GaussianNB, MLPClassifier, and the DecisionTreeCLassifier. 
for i in range(len(classifiers)):
    print(f'\t{classifier_names[i]} : {np.mean(classifier_temp[i])}')
```
